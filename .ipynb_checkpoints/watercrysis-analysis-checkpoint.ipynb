```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
level_df = pd.read_csv('data/chennai_reservoir_levels.csv', index_col = 'Date')
rain_df = pd.read_csv('data/chennai_reservoir_rainfall.csv', index_col = 'Date')
level_df.index = pd.to_datetime(level_df.index, format = '%d-%m-%Y')
rain_df.index = pd.to_datetime(rain_df.index, format = '%d-%m-%Y')
```

```python
level_df.head()
```

```python
rain_df.head()
```

```python
def trend_plot(df, title):
    plt.figure(figsize = (20, 6))
    plt.title(title)
    sns.lineplot(data = df)
    plt.xlabel('Year')
```

```python
trend_plot(df = level_df, title = 'Water Level in Reservoirs')
```

```python
df = pd.DataFrame()
df['Combined_Level'] = level_df['POONDI'] + level_df['CHOLAVARAM'] + level_df['REDHILLS'] + level_df['CHEMBARAMBAKKAM']
df.index = level_df.index
trend_plot(df = df['Combined_Level'], title = 'Combined Water Level')
```

##### Inference:
1. From the graph 'Combined Water Level' we can infer that every year water in reservoirs get replenished roughly between October and December (monsoon season).
2. There seems to have been a water scarcity phase in 2004. This corresponds to our knowledge of the Boxing Day Tsunami in the Indian ocean.
3. We can also infer that Chennai could have suffered a water shortage in 2014 - 2015 but was saved due to water in Redhills and Chembarambakkam.
4. This year in 2019 however we can view that there is a water shortage throughout all 4 reservoirs.
5. The sharp spike in water levels in 2015 is due to the flood that occured then.

```python
trend_plot(df = rain_df['POONDI'], title = 'Rainfall in Poondi')
trend_plot(df = rain_df['CHOLAVARAM'], title = 'Rainfall in Cholavaram')
trend_plot(df = rain_df['REDHILLS'], title = 'Rainfall in Redhills')
trend_plot(df = rain_df['CHEMBARAMBAKKAM'], title = 'Rainfall in Chembarambakkam')
```

```python
df['Rainfall_Combined'] = rain_df['POONDI'] + rain_df['CHOLAVARAM'] + rain_df['REDHILLS'] + rain_df['CHEMBARAMBAKKAM']
trend_plot(df = df['Rainfall_Combined'], title = 'Combined Rainfall')
```
